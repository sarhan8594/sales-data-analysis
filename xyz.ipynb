```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('superstore_sales.csv', encoding='utf-8')
```

```python
import seaborn as sns
```

profit by state


```python
# Group the data by 'State' and calculate the total profit for each state
profit_by_state = df.groupby('State')['Profit'].sum().reset_index()
```

```python
# Create a bar chart for profit by states
plt.figure(figsize=(12, 6))
plt.bar(profit_by_state['State'], profit_by_state['Profit'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.title('Total Profit by State')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

profit by year

```python
# Extract year from the 'Order Date' column
df['Year'] = pd.to_datetime(df['Order Date']).dt.year
```

```python
# Group the data by 'Year' and calculate the total profit for each year
profit_by_year = df.groupby('Year')['Profit'].sum().reset_index()
```

```python
# Create a bar chart with years on the x-axis and profit values on the y-axis
plt.figure(figsize=(10, 6))
plt.bar(profit_by_year['Year'], profit_by_year['Profit'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Total Profit by Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
